```python
import pandas as pd
import numpy as np
import seaborn as sb 
titanic = sb.load_dataset('titanic')
titanic
#titanic.to_csv('titanic_new.csv')
```

```python
titanicFilledNa = titanic.fillna(titanic.mean())
#titanicFilledNa.to_csv('titanicFilledNa.csv')
```

```python
# 3
titanic['survived'].sum() # 생존
titanic[titanic['survived']==0]['survived'].count() # 사망
```

```python
# 4
titanic.pivot_table('survived','pclass', aggfunc='mean')

# 5
titanic.pivot_table('survived','sex', aggfunc='mean')

# 6
titanic['age_class'] = pd.qcut(titanic.age,5,labels=['child','young','middle','old','very old'] )
titanic
titanic.pivot_table('survived','age_class', aggfunc='mean')

# 7
titanic.pivot_table('survived','sibsp', aggfunc='mean')

# 8
titanic.pivot_table('fare','pclass', aggfunc='mean')
```
