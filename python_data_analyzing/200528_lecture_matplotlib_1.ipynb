```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### 선그래프

```python
a = np.array([10,14,19,20,25])
print(a)
plt.plot(a)

a2 = np.arange(20)
print(a2)
plt.plot(a2)

plt.show() # 새 프레임에 그래프를 출력
```

```python
# 1차 방정식의 그래프 : f(x) = a*x + b , 직선의 방정식
x = np.arange(-4.5,5,0.5)
y = 2*x + 0
print(x,y)
plt.plot(x,y)
plt.show()
```

```python
# 2차 방정식의 그래프 : f(x) = a*x^2 + b, 포물선의 방정식
x = np.arange(-4.5,5,0.5)
y = 2*x**2
print(x,y)
plt.plot(x,y)

plt.plot(x,4*x)
plt.plot(x,-3*x)
plt.show()
```

### 로그플롯(log plot)

```python
print(np.log(100)) # 자연로그,  e: 2.71828
print(np.exp(4.605170185988092))  # 100
# log(100) = 4.605170185988092
# 100 =  e**4.605170185988092
```

```python
df = pd.read_csv('transcount.csv')
df = df.groupby('year').agg(np.mean)
df
```

```python
years = df.index.values 
counts = df['trans_count'].values
print(type(years),type(counts))
print(years,counts)
plt.plot(years,counts) # 원본 데이터로 그래프를 출력 , 
                       # 올바른 정보(변화량)를 알 수가 없다
                       # 1970~2000년 사이의 변화량이 매우 작음
plt.show()
```

```python
# y 값을 log변환하여 그래프로 출력
plt.plot(years,np.log(counts)) # 자연로그, e:2.71828
                               # 1970~2000년 사이의 변화량이 분명해짐
plt.show()
```

```python
print(np.log1p(0))  # 1을 더한 후 log변환 ,log(0)이 -inf(-무한대)가 되지 않도록
print(np.log(1))    
# log(0) => -inf
# log(1) => 0

# np.log1p(x) = np.log(0 + 1)
```

```python
### 로그 플롯
years = df.index.values
counts = df['trans_count'].values

plt.semilogy(years,counts,'o')  # semilogy함수는 y 값을 로그변환하여 그래프로 출력

# np.polyfit(x,y,...):x,y값을 사용하여 오차가 가장 적은 직선의 기울기와 y절편을구한다
#                     예측 방정식(H(x) = w*x + b)을 얻는다
poly = np.polyfit(years,np.log(counts),deg=1) # deg=1: x는 1차원
print(poly)  # [ 3.61559210e-01 -7.05783195e+02] ,회귀 계수
# H(x) = 3.61559210e-01*x -7.05783195e+02 ,예측 방정식

plt.semilogy(years,np.exp(np.polyval(poly,years))) # 노란색 직선 ,예측 함수의 그래프

plt.show()

np.exp(np.polyval(poly,years)) # polyfit()으로 얻은 poly내의 기울기와 y절편을 이용하여
# x값을 넣어 예측값(H(x), y축의 값)을 구한다
```

### 분산 플롯(scatter plot): scatter()함수, 버블 차트

```python
df = pd.read_csv('transcount.csv')
df = df.groupby('year').agg(np.mean)

gpu = pd.read_csv('gpu_transcount.csv')
gpu = gpu.groupby('year').agg(np.mean)

df = pd.merge(df,gpu,how='outer',left_index=True,right_index=True)
df = df.replace(np.nan,0) # 결측치를 0으로 변경

years = df.index.values     # x값
counts = df['trans_count'].values
cnt_log = np.log(counts)    # 로그변환한  y값
gpu_counts = df['gpu_trans_count'].values # 버블의 크기

plt.scatter(years,cnt_log, c = 200*years,
           s =20 + 200*gpu_counts/gpu_counts.max(),
           alpha=0.5 ) # c:컬러 ,s:버블의 크기,alpha:투명도
plt.grid(True)
plt.xlabel('years')
plt.ylabel('trans_count')
plt.title('trans/gpu counts pers years')

# # 그래프 한글출력 설정(Windows)
# import matplotlib

# matplotlib.rcParams['font.family'] = 'Malgun Gothic'   # '맑은 고딕'으로 설정 
# matplotlib.rcParams['axes.unicode_minus'] = False      # minus숫자 표시

# plt.title('tran/gpu 연간판매량')
plt.show()
```

```python
# 범례(Legend),주석(annotaion), 예측 방정식의 직선 그래프를 함께 출력
df = pd.read_csv('transcount.csv')
df = df.groupby('year').agg(np.mean)

gpu = pd.read_csv('gpu_transcount.csv')
gpu = gpu.groupby('year').agg(np.mean)

df = pd.merge(df,gpu,how='outer',left_index=True,right_index=True)
df = df.replace(np.nan,0) # 결측치를 0으로 변경

years = df.index.values     # x값
counts = df['trans_count'].values
cnt_log = np.log(counts)    # 로그변환한  y값
gpu_counts = df['gpu_trans_count'].values # 버블의 크기

## 추가 코드

# 예측 방정식의 직선 그래프 출력
poly = np.polyfit(years,cnt_log,deg=1)
plt.plot(years,np.polyval(poly,years),label='Poly Fit')

# annotaion(주석 달기)
gpu_start = gpu.index.values.min() # 1997
y_ann = np.log(df.at[1997,'trans_count']) # 1997년 trans_count 로그 값
# print(gpu_start,y_ann) # 1997 15.913528485217045
ann_str = 'First GPU\n%d'%gpu_start
# print(ann_str)
plt.annotate(ann_str,xy=(gpu_start,y_ann),
             arrowprops=dict(arrowstyle='->'),
             xytext=(-22,+35),textcoords='offset points')

plt.scatter(years,cnt_log, c = 200*years,
           s =20 + 200*gpu_counts/gpu_counts.max(),
           alpha=0.5, label='Scatter Plot' ) # c:컬러 ,s:버블의 크기,alpha:투명도
plt.grid(True)
plt.xlabel('years')
plt.ylabel('trans_count')
plt.title('trans/gpu counts pers years')
plt.legend(loc='upper left')
plt.show()
```

```python
# 지연(lag) 플롯 : time seires data(시계열 데이터) 분석

from pandas.plotting import lag_plot

# 지연값 : 1년
# x축 : 기준년도의 tans_count의 로그값
# y축 : 기준년도 + 1의 tans_count의 로그값
lag_plot(np.log(df['trans_count']))
plt.grid()
plt.show()
```
