#### DataFrame 의 요소 접근하기 : 읽기와 수정
    * (1) df.loc[행번호,열]: 명시적인index
    * (2) df.iloc[행번호, 열]: 함묵적인 index, 파이썬 스타일, int 형 0 부터 시작
    * df.ix[] : 최신 버전에서 삭제됨

```python
# df.loc[]  : 행의 인덱스, 칼럼명 을 기준으로 dataFrame 뽑기
df.loc[3, 'Country']  # 3번 행, Country Column(0번째)
df.loc[:3, 'Country'] # start:end, end-1 까지가 아님
a = df.loc[:3, ['Country', 'CountryID', 'Continent']]
```

```python
# df.iloc[]  : 행의 인덱스, 칼럼의 인덱스를 기준으로 dataFrame 뽑기
df.iloc[3,0]  
df.iloc[:3, 0]
df.iloc[0, -1]
df.iloc[:,:-1] # 마지막 칼럼 제외
```

### 통계 기초
    평균(mean) : 중심 성향에 대한 추정량을 계산할 때 사용
    편차 (Deviation) : 관측값과 평균값과의 차이, (데이터 값 - 평균), 양수나 음수의 값을 갖는다
    분산 (Variance) : 편차의 제곱의 합의 평균, 음수값이 양수로 더해진다, 실제 값보다 너무 멀어져있게 된다
    표준편차(standard Deviation) : 분산에 루트(제곱근)를 씌운 값

```python
df.mad()  # Mean Absolute Deviation 평균 절대 편차
```

```python
df.mean()
```

```python
df.median()  # 중위수 (2사분위수)
```

```python
df.sum()
df.max()
df.min()
df.mode()  # 최빈값 : 최고로 빈도가 높은 값
```

```python
df.var() # 분산
```

```python
df.std() # 표준편차
```

```python
df.skew() # 비대칭도
df.kurt() # 첨도
df.describe() # 요약
```

```python
score_table1 = {'영어':[80, 100, 82],
               '수학':[60, 95, 78]}
score_table2 = {'영어':[87, 90, 100],
               '수학':[70, 86, 100]}

df1 = pd.DataFrame(score_table1)
df2 = pd.DataFrame(score_table2)
new_df1 = df1.append(df2)  # 행의 인덱스가 012 012 와 같이 중복으로 append 되어 문제..
new_df1   # 012012

new_df1 = df1.append(df2, ignore_index=True) # 행 인덱스가 중복되지 않음
new_df1   # 012345

new_df3 = df1.append(df1, ignore_index=True) # 행 인덱스가 중복되지 않음
new_df3   # 012345
```

```python
# 행(수직) 합치기
pd.concat([df1, df2], axis=0, ignore_index=True)  # == df1.append(df2, ignore_index=True)
```

```python
# 열(수평) 로 합치기 : concat(), merge(), join()
# concat()
pd.concat([df1, df2], axis=1)

# merge()
```
